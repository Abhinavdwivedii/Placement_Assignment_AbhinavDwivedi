## To test the hypothesis that the mean scores of all the students are the same, we can use a one-way analysis of variance (ANOVA) test. If the test indicates that the mean scores are significantly different, we can further determine the student with the highest score.

```python
from scipy.stats import f_oneway

# Ho: the mean scores of the students are significantly same.
# H1: the mean scores of the students are significantly different.

# Define the scores for each student
karan_scores = [85, 90, 92]
deepa_scores = [70, 80, 85]
karthik_scores = [90, 85, 88]
chandan_scores = [75, 70, 75]
jeevan_scores = [95, 92, 96]

# Perform one-way ANOVA
statistic, p_value = f_oneway(karan_scores, deepa_scores, karthik_scores, chandan_scores, jeevan_scores)

# Print the test statistic and p-value
print("Test Statistic:", statistic)
print("p-value:", p_value)

if p_value < 0.05:
    print('Reject the null hypothesis')
else:
    print('Accept the null hypothesis')

# Determine the student with the highest score
students = ['karan', 'deepa', 'karthik', 'chandan', 'jeevan']
scores = [karan_scores, deepa_scores, karthik_scores, chandan_scores, jeevan_scores]
highest_score_student = students[scores.index(max(scores))]

# Print the student with the highest score
print("Student with the highest score:", highest_score_student)

```

## Therefore we can conclude that the mean of the candidates of all the exams are not same.
